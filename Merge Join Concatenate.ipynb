```python
import numpy as np
import pandas as pd
```

### CONCATENATION

```python
#Concatenation
```

```python
dataset1 = {
    "A" : ["A1","A2","A3","A4"],
    "B" : ["B1","B2","B3","B4"],
    "C" : ["C1","C2","C3","C4"],
    "D" : ["D1","D2","D3","D4"]
}
dataset2 = {
    "A" : ["A5","A6","A7","A8"],
    "B" : ["B5","B6","B7","B8"],
    "C" : ["C5","C6","C7","C8"],
    "D" : ["D5","D6","D7","D8"]
}

df1 = pd.DataFrame(dataset1,index = [1,2,3,4])
df2 = pd.DataFrame(dataset2,index = [5,6,7,8])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis = 1)
```

### JOIN

```python
#JOIN
```

```python
dataset1 = {
    "A" : ["A1","A1","A3","A4"],
    "B" : ["B1","B2","B3","B4"]
}
dataset2 = {
    "X" : ["X1","X1","X3"],
    "Y" : ["Y1","Y2","Y3"]
}

df1 = pd.DataFrame(dataset1,index = [1,2,3,4])
df2 = pd.DataFrame(dataset2,index = [1,2,3])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2,how = "right")
```

```python
df2.join(df1)
```

### MERGE

```python
#MERGE
```

```python
dataset1 = {
    "A" : ["A1","A2","A3"],
    "B" : ["B1","B2","B3"],
    "Key" : ["K1","K2","K3"]
}
dataset2 = {
   "X" : ["X1","X2","X3","X4"],
    "Y" : ["Y1","Y2","Y3","Y4"],
    "Key" : ["K1","K2","K5","K4"] 
}
df1 = pd.DataFrame(dataset1,index = [1,2,3])
df2 = pd.DataFrame(dataset2,index = [1,2,3,4])
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on = "Key")
```

```python
pd.merge(df1,df2)
```
